```python
%matplotlib inline
import sympy as sym
t, sigma, omega = sym.symbols('t sigma omega', real=True)#definir variables simbólicas
s = sigma + 1j*omega  #sobre carga operador para definir parte imaginaria
x = sym.exp(s*t)#función exponencial simbólica desde sympy
x

class rect(sym.Function): #clase/objeto tipo sympy

    @classmethod #decorador objeto
    def eval(cls, arg):
        return sym.Heaviside(arg + sym.S.Half) - sym.Heaviside(arg - sym.S.Half)
```

Ejercicio

Pruebe otros valores de omega y sigma para crear señales con amplitud creciente/decreciente/constante y diferentes frecuencias angulares.

CRECIENTE: aumento el sigma aumenta la amplitud, aumento el omega aumenta la frecuencia, en este caso para que la funcion sea creciente ambos parametros deben ser positivos, como se muestra en el resultado de la grafica

```python
y = x.subs({omega: -7, sigma : 0.2}) #sustituir variables omega y sigma por valores numéricos, omega + y sigma+
print(r'x(t)=',x)
#se gráfica desde el paquete plot de sympy por ser objetos de variable simbólica
sym.plot(sym.re(y), (t, 0, 2*sym.pi), ylabel=r'Re{$e^{st}$}',ylim=[-10,10])
sym.plot(sym.im(y), (t, 0, 2*sym.pi), ylabel=r'Im{$e^{st}$}',ylim=[-10,10])
```

Decreciente, para esta interpretacion el parametro omega debe ser positivo y el sigma negativo, mientras mas se aleje de cero su decrecimiento se vera mas pronunciado y se dara en un menor tiempo

```python
y = x.subs({omega: -5, sigma : -0.3}) #sustituir variables omega y sigma por valores numéricos
print(r'x(t)=',x)
#se gráfica desde el paquete plot de sympy por ser objetos de variable simbólica
sym.plot(sym.re(y), (t, 0, 2*sym.pi), ylabel=r'Re{$e^{st}$}',ylim=[-2,2])
sym.plot(sym.im(y), (t, 0, 2*sym.pi), ylabel=r'Im{$e^{st}$}',ylim=[-2,2])
```

Constante:Para que sea constante el sigma tiene que ser cero o se hace constante si esta lo mas cerca posible de cero

```python
y = x.subs({omega: -5, sigma : 0}) #sustituir variables omega y sigma por valores numéricos
print(r'x(t)=',x)
#se gráfica desde el paquete plot de sympy por ser objetos de variable simbólica
sym.plot(sym.re(y), (t, 0, 2*sym.pi), ylabel=r'Re{$e^{st}$}',ylim=[-2,2])
sym.plot(sym.im(y), (t, 0, 2*sym.pi), ylabel=r'Im{$e^{st}$}',ylim=[-2,2])
```

Compruebe la propiedad de selectividad, para un impulso de Dirac desplazado  δ$(t−τ)$  y probar los resultados modificando el ejemplo anterior.
Definimos otra funcion delta_T, en el que usamos la ecuacion de DiracDelta con el argumento menos tau, la cual llamamos en la integracion

```python
#f(t-\tau)
#definir tau
tau = sym.symbols('tau', real=True)
g = sym.Function('g')(t) # g(t)
delta_T = sym.DiracDelta(t-tau) # delta(t-tau)
delta_T
sym.integrate(delta_T*g, (t, -sym.oo, sym.oo))
```

Ejercicio

Use la función  rect(t)  para construir una función  $cos(ωt)$  con  ω=2 , la cual se enciende en  t=−12  y se apaga en  t=+12 .

```python
omega=2
c = sym.cos(omega*t) * rect(t)
sym.plot(c, (t, -1, 1), ylim=[-0.2, 1.2], ylabel=r'rect$(t)$');
```

Ejercicio

Genere un rectificador de onda completa utilizando la función  sgn(t)

```python
omeg=2
a=sym.cos(omeg*t)*sym.sign(sym.cos(omeg*t))
sym.plot(a, (t, -2, 2), ylim=[-1.2, 1.2], ylabel=r'sgn$(t)$');
```
